````python
x = df.drop(columns = 'label', axis = 1)
y = df['label']
````

Splitting Train test model

````python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=2)
````

````python
print(x.shape,X_train.shape,X_test.shape)
````

````python
#Standardize the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
````

````python
print(X_train_std)
````

````python
#Building Neural Network
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
````

Flattern the features into array for NN model

````python
#Setting up the layers of NN
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)), #30 columns - Input layer
    keras.layers.Dense(20,activation='relu'), # Hidden Layer
    keras.layers.Dense(2,activation='sigmoid') #number of labels - Output layer
])
````

````python
#compiling the neural network
model.compile(optimizer='adam',
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy']
)
````

````python
#training neural network
history = model.fit(X_train_std,Y_train,validation_split=0.1,epochs=10)
````

````python
#Visualizing the data
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training_data','validation_data'],loc='lower right')
plt.show()
````

````python
#Visualizing the data
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training_data','validation_data'],loc='upper right')
plt.show()
````

````python
#Accuracy of the model on test data
loss, accuracy = model.evaluate(X_test_std,Y_test)
print("Test Loss:",loss)
print("Test Accuracy:",accuracy)
````
